```python

time_data = [2, 4, 6, 1.5, 3, 5, 7, 2.5]


low_study = []
moderate_study = []
high_study = []


for hours in time_data:
    if hours < 3:
        low_study.append(hours)
    elif 3 <= hours <= 5:
        moderate_study.append(hours)
    else:  # hours > 5
        high_study.append(hours)


print("Low study time:", low_study)
print("Moderate study time:", moderate_study)
print("High study time:", high_study)


print("Number of days with low study time:", len(low_study))
print("Number of days with moderate study time:", len(moderate_study))
print("Number of days with high study time:", len(high_study))
```

```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)  # Recursively sum nested lists
        else:
            total += element  # Add the number directly
    return total

# Test the function
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total sum:", result)
```

```python
def generate_permutations(s):
    if len(s) == 0:
        return ['']
    permutations = []
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            permutations.append(char + p)
    return list(set(permutations))

print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total directory size:", calculate_directory_size(directory_structure), "KB")
```

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1, 2, 5], 11))
```

```python
def longest_common_subsequence(s1, s2):
    dp = [[0] * (len(s2)+1) for _ in range(len(s1)+1)]
    for i in range(len(s1)):
        for j in range(len(s2)):
            if s1[i] == s2[j]:
                dp[i+1][j+1] = dp[i][j] + 1
            else:
                dp[i+1][j+1] = max(dp[i][j+1], dp[i+1][j])
    return dp[-1][-1]

print(longest_common_subsequence("abcde", "ace"))
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][capacity]

print(knapsack([1, 3, 4, 5], [1, 4, 5, 7], 7))
```
